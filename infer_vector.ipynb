## Imports:

```
#import required modules
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_hub as hub
from sklearn import preprocessing
import logging
pip install docx2txt
logging.getLogger('tensorflow').disabled = True #OPTIONAL - to disable outputs from Tensorflow
import glob 
import re
import docx2txt 
```

## Create sentence embeddings

```
#download elmo- pretrained model
url = "https://tfhub.dev/google/elmo/2"
embed = hub.Module(url)
```

```
#take in all document names
doc_name_list=(glob.glob("*.docx"))
print(doc_name_list)

processed=[]
text=""
matches=[]
#read all documents one after another (the user will have to make changes here that suite their documnet type)
#by default name of file containing document is considered name for that document which are stored in variable 'matches'
for i in range(len(doc_name_list)):
  text+=docx2txt.process(doc_name_list[i])	
  text+="\n"
  matches.append(doc_name_list[i])
final_doc=[]		
for j in range(len(matches)-1):
  final_doc.append(text[text.index(matches[j]):text.index(matches[j+1])])
final_doc.append(text[text.index(matches[-1]):])
text=final_doc
for j in range(len(text)):
  text[j]=str(re.sub(r'[\t|\n|\xa0]',r' ',text[j]))
```

```
#write names of documents to text file which can be used further
f=open('doc_names.txt','w')
for j in range(len(matches)):
  f.write(str(matches[j]))
  f.write("\n")
f.close()
```

```
#infer vectors for your documents using pretrained model(takes a lot of time)
#Do not make any changes except variable 'siz' because that is number of documents given to model at a time.(size of batch)
#If it shows any error while inferring just reduce the variable 'siz'- should be a positive number
sentences=text[:]
fin_vecs=[]
num=0
siz=15
while(num<len(sentences)-siz):
  batches=sentences[num:num+siz]
  embeddings = embed(
    batches,
    signature="default",
    as_dict=True)["default"]
  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    x = sess.run(embeddings)
 
  for each in x:
    fin_vecs.append(each)
  num+=siz
  print(num,len(fin_vecs))  
print(num,len(fin_vecs))
print(len(sentences[num:]))
batches=sentences[num:]
embeddings = embed(
  batches,
  signature="default",
  as_dict=True)["default"]
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  sess.run(tf.tables_initializer())
  x = sess.run(embeddings)
print(len(x))
for each in x:
  fin_vecs.append(each)
print(len(fin_vecs))
```

```
#store the vectors inferred(size 1024*8 bytes)
f=open('embeds.txt','w')
for i in fin_vecs:
  for j in i:
    f.write(str(j)+" ")
  f.write("end\n")
f.close()
```
